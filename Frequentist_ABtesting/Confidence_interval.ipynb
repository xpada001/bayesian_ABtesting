```python
import numpy as np
from scipy.stats import norm, t
```

```python
np.random.seed(1)

N = 1000
mu = 5
sigma = 2
X = np.random.randn(N)*sigma+mu
```

```python
mu_h = np.mean(X)
sigma_h = np.std(X, ddof=1)
z_left = norm.ppf(0.025)
z_right = norm.ppf(0.975)

t_left = t.ppf(0.025, df = N-1)
t_right = t.ppf(0.975, df = N-1)
```

```python
lower_interval = mu_h + z_left*sigma_h/np.sqrt(N)
higher_interval = mu_h + z_right*sigma_h/np.sqrt(N)
print(("mu: {}, lower internal: {}, higher interval: {}").format(mu_h, lower_interval, higher_interval))
```

```python
lower_interval_t = mu_h + t_left*sigma_h/np.sqrt(N)
higher_interval_t = mu_h + t_right*sigma_h/np.sqrt(N)
print(("mu: {}, lower internal: {}, higher interval: {}").format(mu_h, lower_interval_t, higher_interval_t))
```

```python
#now we test the definition of confidence interval, i.e. return the rate of our true mean appear in the interval

def sample_process(ite):
  res = 0
  for i in range(ite):
    X = np.random.randn(N)*sigma+mu
    mu_h = np.mean(X)
    sigma_h = np.std(X, ddof=1)
    t_left = t.ppf(0.025, df = N-1)
    t_right = t.ppf(0.975, df = N-1)
    lower_interval_t = mu_h + t_left*sigma_h/np.sqrt(N)
    higher_interval_t = mu_h + t_right*sigma_h/np.sqrt(N)
    res += mu >= lower_interval_t and mu <= higher_interval_t

  return res / ite
```

```python
sample_process(10000)
```
